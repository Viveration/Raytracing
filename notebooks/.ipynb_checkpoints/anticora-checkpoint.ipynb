```python
import numpy as np
import plotly.graph_objects as go
import pandas as pd
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
from typing import List, Union
import scipy.optimize as spo
from collections import Counter
```

```python
def slice_triangles(z, n, i, j, k, l):
    """Create the triangles of a single slice"""
    return [[z, j, i], [i, j, l], [l, j, k], [k, n, l]]


def cylinder_mesh(r, xs, ys, zs, h, n_slices=40):
    """Create a cylindrical mesh"""
    theta = np.linspace(0, 2 * np.pi, n_slices + 1)
    x = xs + r * np.cos(theta)
    y = ys + r * np.sin(theta)
    z1 = zs + 0 * np.ones_like(x)
    z2 = (zs + h) * np.ones_like(x)
# index of the final point in the mesh
    n = n_slices * 2 + 1

    # build triangulation
    triangles = []
    for s in range(1, n_slices + 1):
        j = (s + 1) if (s <= n_slices - 1) else 1
        k = j + n_slices if (s <= n_slices - 1) else n_slices + 1
        l = s + n_slices
        triangles += slice_triangles(0, n, s, j, k, l)
    triangles = np.array(triangles)

    # coordinates of the vertices
    x_coords = np.hstack([xs, x[:-1], x[:-1], xs])
    y_coords = np.hstack([ys, y[:-1], y[:-1], ys])
    z_coords = np.hstack([zs, z1[:-1], z2[:-1], (zs + h)])
    vertices = np.stack([x_coords, y_coords, z_coords]).T

    return vertices, triangles, x, y, z1, z2


def cylinder_traces(r, xs, ys, zs, h, n_slices=40, show_mesh=False, n_sub=4, surface_kw=dict(), line_kw=dict()):
    """

    Parameters
    ----------
    r : `float`
        Радиус цилиндра.
    xs : `float`
        Набор х-координат.
    ys : `float`
        Набор y-координат.
    zs : `float`
        Набор z-координат.
    h : `float`
        Высота цилиндра.
    n_slices : `int`, optional
        Разбиение по углу, by default 40
    show_mesh : `bool`, optional
        Показать сеть, by default False
    n_sub : `int`, optional
        Разбиение по высоте, by default 4
    surface_kw : `dict()`, optional
        Дополнительные аргументы поверхности, by default dict()
    line_kw : `dict()`, optional
        Дополнительные аргументы линии, by default dict()

    Returns
    -------
    `array_like`
        Набор координат для изображения цилиндра
    """ 

    vertices, triangles, x, y, z1, z2 = cylinder_mesh(
        r, xs, ys, zs, h, n_slices)
    surface = go.Mesh3d(
        x=vertices[:, 0], y=vertices[:, 1], z=vertices[:, 2],
        i=triangles[:, 0], j=triangles[:, 1], k=triangles[:, 2],
        **surface_kw)

    traces = [surface]
    if not show_mesh:
        return traces

    line_kw.setdefault("showlegend", False)
    # horizontal mesh lines
    zsubs = np.linspace(zs, zs + h, n_sub + 1)
    for zc in zsubs:
        traces.append(go.Scatter3d(x=x, y=y, z=zc *
                      np.ones_like(x), mode="lines", **line_kw))
    # vertical mesh lines
    for _x, _y in zip(x, y):
        traces.append(go.Scatter3d(x=[_x, _x], y=[_y, _y], z=[
                      zs, zs + h], mode="lines", **line_kw))
    return traces

def show_trajectory(data, index, fiber1, fiber2, cap=None):
    dots = np.array(data['trajectories'][index])
    if not cap:
        cap = int(data['reflections'][index])
    fig = go.Figure()
    fig.add_traces(
        cylinder_traces(fiber1.core_r, 0, 0, 0, dots[cap-1][2], 100, n_sub=8, line_kw={"line_color":"#202020", "line_width": 3})
    )
    fig.add_traces(
        cylinder_traces(fiber2.core_r, 0, 0, 0, dots[cap-1][2], 100, n_sub=8, line_kw={"line_color":"#303030", "line_width": 3})
    )
    fig.update_traces(opacity=0.15, selector=dict(type='mesh3d'))
    fig.add_traces(data=go.Scatter3d(
        x=dots[:cap, 0], y=dots[:cap, 1], z=dots[:cap, 2],
        marker=dict(
            size=3,
            color='black',
            colorscale='Viridis',
        ),
        line=dict(
            color='darkblue',
            width=2
        )
    ))
    fig.update_layout(
        autosize=False,
        width=800,
        height=800,)
    fig.show()
```

```python
class Fiber_cylinder:
    def __init__(self, core_r = 0, clad_r = 0, core_n = 0, clad_n = 0, z_max = 0, diffusion = None):
        self.diffusion = diffusion
        self.core_r = core_r
        self.clad_r = clad_r
        self.core_n = core_n
        self.clad_n = clad_n
        self.z_max = z_max
    def set_geometry(self, core_r, clad_r, z): #радиусы принимаются в метрах
        self.core_r = core_r
        self.clad_r = clad_r
        self.z_max = z
    def set_refr(self, core_n, clad_n):
        self.core_n = core_n
        self.clad_n = clad_n
    def radius(self):
        return self.core_r, self.clad_r
    def refr(self):
        return self.core_n, self.clad_n
    def find_normal(self, intersection_point):
        x, y = intersection_point[:2]
        result = -1 * np.array([x, y, 0])/np.sqrt(x**2 + y**2)
        return result # Возвращает нормаль к поверхности, направленную к центру цилиндра
    
```

```python
class Ray_cylinder:
    def __init__(self, azimut = 0, latitude = 0, startpoint = [0,0,0]): #все углы принимаются в радианах
        self.cdf = None
        self.angles = None
        self.azimut = azimut
        self.latitude = latitude
        self.startpoint = startpoint
        self.vector = [np.sin(latitude)*np.cos(azimut),
                      np.sin(latitude)*np.sin(azimut), 
                      np.cos(latitude)]
    
    def set_values(self, azimut, latitude, startpoint):
        self.azimut = azimut
        self.latitude = latitude
        self.startpoint = startpoint
        self.vector = [np.sin(latitude)*np.cos(azimut),
                      np.sin(latitude)*np.sin(azimut), 
                      np.cos(latitude)]
    def calculate_angles_from_vector(self, vector):
        latitude = np.arccos(vector[2])
        if latitude == 0:
            return [0, 0]
        cos = vector[0] / np.linalg.norm(vector[:2])
        sin = vector[1] / np.linalg.norm(vector[:2])
        if sin > 0:
            azimut = np.arccos(cos)
        else:
            azimut = -np.arccos(cos)
        return [azimut, latitude]
    
    def calculate_intersection(self, core_radius):
        if self.latitude == 0:
            print('zero')
            return [*self.startpoint[:2], -1]
        x0, y0, z0 = self.startpoint
        phi, alpha = self.azimut, self.latitude
        R = core_radius
        gamma = np.cos(phi)*x0 + np.sin(phi)*y0
        x = x0 + np.cos(phi) * (-gamma + np.sqrt(gamma**2 - x0**2 - y0**2 + R**2))
        y = y0 + np.sin(phi) * (-gamma + np.sqrt(gamma**2 - x0**2 - y0**2 + R**2))
        z = z0 + np.cos(alpha) / np.sin(alpha) * (-gamma + np.sqrt(gamma**2 - x0**2 - y0**2 + R**2))

        return [x, y, z]
    
    def calculate_reflection(self, point, normal):
        angle  = np.dot(normal, self.vector)
        proection = normal * angle
        reflected_vector = -2 * proection + self.vector
        return *self.calculate_angles_from_vector(reflected_vector), angle
    
    def generate_startpoint(self, radius):
        phi = np.random.random() * np.pi
        r = np.random.random() * radius
        coords = [r*np.cos(phi), r*np.sin(phi)]
        self.startpoint = [*coords, 0]
        return [*coords, 0]
    
    def set_startpoint(self, x = 0, y = 0, z = 0):
        self.startpoint = [x, y, z]
        return self.startpoint
    
    def generate_angles(self, max_latitude = 30):
        latitude = np.deg2rad(np.random.random() * max_latitude)
        azimut = np.deg2rad(np.random.random() * 360)
        self.azimut = azimut
        self.latitude = latitude
        self.vector = [np.sin(latitude)*np.cos(azimut),
                      np.sin(latitude)*np.sin(azimut), 
                      np.cos(latitude)]
        return [latitude, azimut]
    
    def generate_angles_uniform(self, max_latitude = 10, step = 0.1):
        if  self.cdf is None:                                                   # Если кумулятивная функция распределения не посчитана
            n = max_latitude / step                                                # Количество точек разбиения
            self.angles = np.linspace(0, np.deg2rad(max_latitude), int(n))         # Массив углов с осью 
            solid_angles = np.array([2*np.pi*(1-np.cos(x)) for x in self.angles])  # Массив телесных углов
            solid_angles_diff = np.diff(solid_angles)                              # Массив производных телесных углов
            pdf = solid_angles_diff / sum(solid_angles_diff)                       # Функция плотности вероятности
            self.cdf = np.array([np.sum(pdf[:i]) for i in range(len(pdf)+1)])      # Кумулятивная функция плотности распределения
        self.azimut = np.deg2rad(np.random.random() * 360)
        self.latitude = self.angles[next(x[0] for x in enumerate(self.cdf) if x[1] > np.random.random())] # Обратная cdf
        self.vector = [np.sin(self.latitude)*np.cos(self.azimut),
                      np.sin(self.latitude)*np.sin(self.azimut), 
                      np.cos(self.latitude)]
        print(np.rad2deg(self.latitude))
        return [self.latitude, self.azimut]
    
    def set_angles(self, azimut, latitude):
        self.azimut = azimut
        self.latitude = latitude
        self.vector = [np.sin(latitude)*np.cos(azimut),
                      np.sin(latitude)*np.sin(azimut), 
                      np.cos(latitude)]
        return [latitude, azimut]
    
    def calculate_trajectory(self, fiber, max_reflection = 1000, angle_elimination = False, output = False):
        z_max = fiber.z_max
        termination_angle = np.arcsin(fiber.clad_n / fiber.core_n)
        Dots = np.zeros((max_reflection, 3))
        reflection_angle = self.latitude
        Angles = np.zeros((max_reflection, 3))
        z_start = self.startpoint[2]
        Dots[0] = self.startpoint
        Angles[0] = [self.azimut, self.latitude, 0]
        for i in range(1, max_reflection):
            intersection_point = self.calculate_intersection(fiber.core_r)
            if intersection_point[2] > z_max:
                intersection_final = intersection_point - self.vector / self.vector[2] * (intersection_point[2] - z_max) 
                Dots[i] = intersection_final
                Angles[i] = [self.azimut, self.latitude, abs(reflection_angle)]
                if output:
                    print('Ray reached z_max.')
                return Dots, np.rad2deg(Angles), i + 1, 'z_max'
            normal = fiber.find_normal(intersection_point)
            azimut, latitude, reflection_angle = self.calculate_reflection(intersection_point, normal)
            if fiber.diffusion:
                azimut += (np.random.random() - 0.5 ) * 2 * fiber.diffusion
                latitude += (np.random.random() - 0.5 ) * 2 * fiber.diffusion
            Dots[i] = intersection_point
            Angles[i] = [azimut, latitude, abs(reflection_angle)]
            if angle_elimination:
                if np.pi / 2 - abs(reflection_angle) < termination_angle:
                    if output:
                        print('Ray terminated, termination angle: ', np.rad2deg(termination_angle),' reflection_angle: ',  np.rad2deg(np.pi / 2 - abs(reflection_angle)))
                    return Dots, np.rad2deg(Angles), i + 1, 'reflection_angle'
            self.set_values(azimut, latitude, intersection_point)
        if output:
            print('Reflections exceeded', max_reflection)
        return Dots, np.rad2deg(Angles), max_reflection, 'max_reflections'
    
def calculate_multiple_trajectories(fiber, trajectory_count = 100, max_reflection = 1000, output = False, angle = 8, startpoints = None, startangles = None):
    trajectories = np.zeros((trajectory_count, max_reflection, 3))
    angles = np.zeros((trajectory_count, max_reflection, 3))
    reflections = np.zeros(trajectory_count)
    termination = ["" for x in range(trajectory_count)] # 0 - max_z, 1 - max_reflection, 2 - reflection angle
    if isinstance(fiber, Fiber_cylinder):
        ray_internal = Ray_cylinder()
    else:
        ray_internal = Ray_cone()
    for i in tqdm(range(trajectory_count), leave = False):
        
        if startpoints:
            ray_internal.set_startpoint(*startpoints[i])
        else:
            if isinstance(fiber, Fiber_cylinder):
                ray_internal.generate_startpoint(fiber.core_r)
            else:
                ray_internal.generate_startpoint(fiber.base_r)
                
        if startangles:
            ray_internal.set_angles(*startangles[i])
        else:    
            ray_internal.generate_angles_uniform(max_latitude = angle)
            
            
        trajectories[i], angles[i], reflections[i], termination[i] = ray_internal.calculate_trajectory(fiber, max_reflection=max_reflection, output = output)
    dictionary = {
        'angles':       angles.tolist(),
        'trajectories': trajectories.tolist(),
        'reflections':  reflections.tolist(),
        'termination':  termination
    }
    return pd.DataFrame(dictionary)
```

```python
class Ray_anti:
    def __init__(self, azimut = 0, latitude = 0, startpoint = [0,0,0]): #все углы принимаются в радианах
        self.cdf = None
        self.angles = None
        self.index = True
        self.azimut = azimut
        self.latitude = latitude
        self.startpoint = startpoint
        self.vector = [np.sin(latitude)*np.cos(azimut),
                      np.sin(latitude)*np.sin(azimut), 
                      np.cos(latitude)]
    
    def set_values(self, azimut, latitude, startpoint):
        self.azimut = azimut
        self.latitude = latitude
        self.startpoint = startpoint
        self.vector = [np.sin(self.latitude)*np.cos(self.azimut),
                      np.sin(self.latitude)*np.sin(self.azimut), 
                      np.cos(self.latitude)]
    def calculate_angles_from_vector(self, vector):
        latitude = np.arccos(vector[2])
        if latitude == 0:
            return [0, 0]
        cos = vector[0] / np.linalg.norm(vector[:2])
        sin = vector[1] / np.linalg.norm(vector[:2])
        if sin > 0:
            azimut = np.arccos(cos)
        else:
            azimut = -np.arccos(cos)
        return [azimut, latitude]
    
    def calculate_intersection(self, core_radius, answer = 'max'):
        if self.latitude == 0:
            return [*self.startpoint[:2], -1]
        x0, y0, z0 = self.startpoint
        phi, alpha = self.azimut, self.latitude
        R = core_radius
        gamma = np.cos(phi)*x0 + np.sin(phi)*y0
        if gamma**2 - x0**2 - y0**2 + R**2 >= 0:
            if answer == 'max':
                x = x0 + np.cos(phi) * (-gamma + np.sqrt(gamma**2 - x0**2 - y0**2 + R**2))
                y = y0 + np.sin(phi) * (-gamma + np.sqrt(gamma**2 - x0**2 - y0**2 + R**2))
                z = z0 + np.cos(alpha) / np.sin(alpha) * (-gamma + np.sqrt(gamma**2 - x0**2 - y0**2 + R**2))
            else:
                x = x0 + np.cos(phi) * (-gamma - np.sqrt(gamma**2 - x0**2 - y0**2 + R**2))
                y = y0 + np.sin(phi) * (-gamma - np.sqrt(gamma**2 - x0**2 - y0**2 + R**2))
                z = z0 + np.cos(alpha) / np.sin(alpha) * (-gamma - np.sqrt(gamma**2 - x0**2 - y0**2 + R**2))
            return [x, y, z]
        else:
            return None
    def calculate_refraction(self, normal, n1, n2):
        v1 = np.array(self.vector)
        cos_i = abs(np.dot(v1, normal))
        sin_i = 1-cos_i**2
        sin_t = sin_i * n1 / n2
        if sin_t > 1:
            print('Error')
            # print('v1:', v1, 'norm:', normal,'start:', self.startpoint, n1, n2)
            return None
        cos_t = np.sqrt(1 - sin_t**2)
        normal = np.array(normal)
        refracted_vector = n1 / n2 * v1 + normal * (n1 / n2 * cos_i - cos_t)
        if refracted_vector[2] == 0:
            print('refr')
        print('v1:', v1, 'v2:', refracted_vector, 'norm:', normal,'start:', self.startpoint, n1, n2)
        return *self.calculate_angles_from_vector(refracted_vector), 0
    
    def full_internal(self, normal, n1, n2):
        cosine = np.dot(normal, self.vector)
        angle = np.arccos(cosine)
        if angle < np.arcsin(n2/n1):
            return False
        else:
            return True
        
    def find_direction(self, point):
        vec = np.array(point) - np.array(self.startpoint)
        if np.dot(self.vector, vec) > 0:
            return True
        else:
            return False
    def calculate_reflection(self, point, normal):
        angle  = np.dot(normal, self.vector)
        proection = normal * angle
        reflected_vector = -2 * proection + self.vector
        if reflected_vector[2] == 0:
            print('refl')
        return *self.calculate_angles_from_vector(reflected_vector), angle
    
    def generate_startpoint(self, radius):
        phi = np.random.random() * np.pi
        r = np.random.random() * radius
        coords = [r*np.cos(phi), r*np.sin(phi)]
        self.startpoint = [*coords, 0]
        return [*coords, 0]
    
    def set_startpoint(self, x = 0, y = 0, z = 0):
        self.startpoint = [x, y, z]
        return self.startpoint
    
    def generate_angles(self, max_latitude = 30):
        latitude = np.deg2rad(np.random.random() * max_latitude)
        azimut = np.deg2rad(np.random.random() * 360)
        self.azimut = azimut
        self.latitude = latitude
        self.vector = [np.sin(latitude)*np.cos(azimut),
                      np.sin(latitude)*np.sin(azimut), 
                      np.cos(latitude)]
        return [latitude, azimut]
    def generate_angles_uniform(self, max_latitude = 10, step = 0.1):
        if self.cdf is None:                                                      # Если кумулятивная функция распределения не
                                                                              # посчитана
            n = max_latitude / step                                           # Количество точек разбиения
            self.angles = np.linspace(0, np.deg2rad(max_latitude), int(n))    # Массив углов с осью 
            solid_angles = np.array([2*np.pi*(1-np.cos(x)) for x in self.angles])  # Массив телесных углов
            solid_angles_diff = np.diff(solid_angles)                         # Массив производных телесных углов
            pdf = solid_angles_diff / sum(solid_angles_diff)                  # Функция плотности вероятности
            self.cdf = np.array([np.sum(pdf[:i]) for i in range(len(pdf)+1)]) # Кумулятивная функция плотности распределения
        self.azimut = np.deg2rad(np.random.random() * 360)
        self.latitude = self.angles[next(x[0] for x in enumerate(self.cdf) if x[1] > np.random.random())] # Обратная cdf
        return [self.azimut, self.latitude]

    def set_angles(self, azimut, latitude):
        self.azimut = azimut
        self.latitude = latitude
        self.vector = [np.sin(latitude)*np.cos(azimut),
                      np.sin(latitude)*np.sin(azimut), 
                      np.cos(latitude)]
        return [latitude, azimut]
    
    def calculate_trajectory(self, fiber1, fiber2, max_reflection = 1000, angle_elimination = True, output = False):
        z_max = fiber1.z_max
        if np.linalg.norm(self.startpoint) < fiber2.core_r:
            self.index = False
        else:
            self.index = True
        termination_angle = np.arcsin(fiber1.clad_n / fiber1.core_n)
        Dots = np.zeros((max_reflection, 3))
        reflection_angle = self.latitude
        Angles = np.zeros((max_reflection, 3))
        indexes = np.zeros(max_reflection)
        z_start = self.startpoint[2]
        indexes[0] = self.index
        Dots[0] = self.startpoint
        Angles[0] = [self.azimut, self.latitude, 0]
        print(self.index)
        for i in range(1, max_reflection-1):
            if not self.index: # Если луч уже находится в сердцевине
                intersection_point = self.calculate_intersection(fiber2.core_r)
                normal = fiber2.find_normal(intersection_point)
                self.index = True
                azimut, latitude, reflection_angle = self.calculate_refraction(normal, fiber2.core_n, fiber1.core_n)
                indexes[i] = self.index
                Angles[i] = [azimut, latitude, abs(reflection_angle)]
                
                if intersection_point[2] > z_max:
                    intersection_final = intersection_point - self.vector / self.vector[2] * (intersection_point[2] - z_max) 
                    Dots[i] = intersection_final
                    
                    if output:
                        print('Ray reached z_max.')
                    return Dots, np.rad2deg(Angles), i + 1, 'z_max', indexes
                Dots[i] = intersection_point
                
                self.set_values(azimut, latitude, intersection_point)
            
            else:    
                intersection_core = self.calculate_intersection(fiber2.core_r, answer='min')
                if intersection_core:
                    normal_core = fiber2.find_normal(intersection_core)
                if intersection_core and self.find_direction(intersection_core) and self.full_internal(normal_core, fiber1.core_n, fiber2.core_n):
                    azimut, latitude, reflection_angle = self.calculate_reflection(intersection_core, normal_core)
                    Angles[i] = [azimut, latitude, abs(reflection_angle)]
                    indexes[i] = self.index
                    if intersection_core[2] > z_max:
                        intersection_final = intersection_core - self.vector / self.vector[2] * (intersection_core[2] - z_max) 
                        Dots[i] = intersection_final
                        if output:
                            print('Ray reached z_max.')
                        return Dots, np.rad2deg(Angles), i + 1, 'z_max', indexes
                    Dots[i] = intersection_core
                    self.set_values(azimut, latitude, intersection_core)
                
                elif intersection_core and self.find_direction(intersection_core) and not self.full_internal(normal_core, fiber1.core_n, fiber2.core_n):
                    azimut, latitude, reflection_angle = self.calculate_refraction(normal_core, fiber1.core_n, fiber2.core_n)
                    Angles[i] = [azimut, latitude, abs(reflection_angle)]
                    self.index = False
                    indexes[i] = self.index
                    if intersection_core[2] > z_max:
                        intersection_final = intersection_core - self.vector / self.vector[2] * (intersection_core[2] - z_max) 
                        Dots[i] = intersection_final
                        if output:
                            print('Ray reached z_max.')
                        return Dots, np.rad2deg(Angles), i + 1, 'z_max', indexes
                    Dots[i] = intersection_core
                    self.set_values(azimut, latitude, intersection_core)
                else:
                    intersection_point = self.calculate_intersection(fiber1.core_r)
                    normal = fiber1.find_normal(intersection_point)
                    azimut, latitude, reflection_angle = self.calculate_reflection(intersection_point, normal)
                    Angles[i] = [azimut, latitude, abs(reflection_angle)]
                    indexes[i] = self.index
                    Dots[i] = intersection_point

                    if intersection_point[2] > z_max:
                        intersection_final = intersection_point - self.vector / self.vector[2] * (intersection_point[2] - z_max) 
                        Dots[i] = intersection_final

                        if output:
                            print('Ray reached z_max.')
                        return Dots, np.rad2deg(Angles), i + 1, 'z_max', indexes
                    
                    self.set_values(azimut, latitude, intersection_point)
                    if angle_elimination:
                        if np.pi / 2 - abs(reflection_angle) < termination_angle and abs(reflection_angle) > 0:
                            if output:
                                print('Ray terminated, termination angle: ', np.rad2deg(termination_angle),' reflection_angle: ',  np.rad2deg(np.pi / 2 - abs(reflection_angle)))
                            return Dots, np.rad2deg(Angles), i + 1, 'reflection_angle', indexes

        if output:
            print('Reflections exceeded', max_reflection)
        return Dots, np.rad2deg(Angles), max_reflection, 'max_reflections', indexes
```

```python
def calculate_multiple_trajectories_anti(fiber1, fiber2, trajectory_count = 100, max_reflection = 1000, output = False, angle = 8, startpoints = None, startangles = None):
    trajectories = np.zeros((trajectory_count, max_reflection, 3))
    angles = np.zeros((trajectory_count, max_reflection, 3))
    indexes = np.zeros((trajectory_count, max_reflection))
    reflections = np.zeros(trajectory_count)
    termination = ["" for x in range(trajectory_count)] # 0 - max_z, 1 - max_reflection, 2 - reflection angle
    for i in tqdm(range(trajectory_count), leave = False):
        ray_internal = Ray_anti()
        if startpoints is not None:
            ray_internal.set_startpoint(*startpoints[i])
        else:
            ray_internal.generate_startpoint(fiber1.core_r)
         
        if startangles is not None:
            ray_internal.set_angles(*startangles[i])
        else:    
            ray_internal.generate_angles_uniform(max_latitude=angle)
            
            
        trajectories[i], angles[i], reflections[i], termination[i], indexes[i] = ray_internal.calculate_trajectory(fiber1, fiber2, max_reflection=max_reflection, output = output)
    dictionary = {
        'angles':       angles.tolist(),
        'trajectories': trajectories.tolist(),
        'reflections':  reflections.tolist(),
        'termination':  termination,
        'indexes':      indexes.tolist()
    }
    return pd.DataFrame(dictionary)
```

```python
fiber_1 = Fiber_cylinder(core_r=62.5e-6, clad_r=660e-6, core_n=1.445, clad_n=1.2, z_max=0.2)
fiber_2 = Fiber_cylinder(core_r=27.5e-6, clad_r=660e-6, core_n=1.438, clad_n=1.445, z_max=0.2)
```

```python
def Extract_finish_data(data):
    data_zm = data[(data['termination'] == 'z_max')].reset_index()
    print(data_zm.shape[0])
    finish_points = [data_zm['trajectories'][i][int(data_zm['reflections'][i]) - 1] for i in range(len(data_zm))]
    for dot in finish_points:
        dot[2] = 0
    finish_angles = [data_zm['angles'][i][int(data_zm['reflections'][i]) - 1][:2] for i in range(len(data_zm))]
    return finish_points, list(np.deg2rad(finish_angles))
```
